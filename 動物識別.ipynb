```python
import gradio as gr
import numpy as np

from keras import models

import cv2
from google.colab.patches import cv2_imshow
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
cnn = models.load_model('draw_model_test_b.h5')
```

```python
dict_animal = {0:'ネコ',1:"サメ",2:"ペンギン",3:"タコ"}
```

```python
def classifierNum(input):
  input = cv2.resize(input, dsize=(28, 28), interpolation=cv2.INTER_CUBIC)
  input = cv2.cvtColor(input, cv2.COLOR_BGR2GRAY)

  input = np.expand_dims(input, axis=-1)
  input = np.expand_dims(input, axis=0)

  input = 255-input

  pred = cnn.predict(input)

  pred_result = {}
  
  for i, p in enumerate(pred[0].tolist()):
    pred_result[dict_animal[i]] = p
  
  print(pred_result)
  

  return pred_result
```

```python
with gr.Blocks() as demo:
  gr.Markdown("""ネコ、サメ、ペンギン、タコ識別""") 
  with gr.Row(): 
    canvas = gr.Paint() 
    pred = gr.Label(num_top_classes=4) 
  btn = gr.Button("Predict") 

  btn.click(fn=classifierNum, inputs=canvas, outputs=pred)

demo.launch(auth=('user1', '1234'), share=True) 
#demo.launch(debug=True)
```
